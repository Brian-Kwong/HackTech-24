```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-distilroberta-v1")
```

```python
sentences = [
    "youtube.com MrBeast 1 v 100 countries challenge",
    "youtube.com How to calculate the area of a cricle",
    "stackoverflow.com Area of a trapezoid"
]
sentence_embeddings = model.encode(sentences)
```

```python
sentence_embeddings
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
depth = 1.7
vagueness = 1.2
# can be done in javascript package
clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=vagueness) # affinity='cosine', linkage='average')
clustering_model.fit(sentence_embeddings)
```
